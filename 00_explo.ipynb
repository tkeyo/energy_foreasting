```python
import pandas as pd
```

```python
df = pd.read_csv("ProductionConsumptionSettlement.csv", delimiter=";", decimal=",")
```

```python
df.info()
```

```python
df = (
    df
    .assign(
        HourUTC=pd.to_datetime(df.HourUTC),
        HourDK=pd.to_datetime(df.HourDK),
    )
)
```

```python
cols = list(df.columns)
cols
```

```python
cols_meta = cols[:3]
cols_meta
```

```python
cols_production = cols[3:cols.index('UnknownProdMWh') + 1]
cols_production
```

```python
# includes exchange - can be negative or positive AND also consumption
cols_other = cols[cols.index('UnknownProdMWh') + 1:]
cols_other
```

```python
df_production = (
    df
    [cols_meta + cols_production]
    # .query("PriceArea == 'DK1'")
    .assign(
        TOTAL_PRODUCTION=df[cols_production].sum(axis=1)
    )
)

df_production[:5]
```

```python
(
    df_production
    # [["HourUTC", "TOTAL_PRODUCTION"]]
    .set_index("HourUTC")
    # .groupby([
    #     pd.Grouper(freq="D"),
    # ])
    # .agg(
    #     DAILY_PRODUCTION=("TOTAL_PRODUCTION", "sum")
    # )
    .to_parquet("Production.parquet")
)
```

```python
df_others = (
    df
    [cols_meta + cols_other]
    .set_index("HourUTC")
)

df_others.to_parquet("Others.parquet")

df_others[:5]
```
